## Bar charts for coping variables by age, gender and degree

```python
#take SSSI_avg as an example to try out bar charts by age, gender and degree 
sssi2 = df[['SSSI_avg','Age']]
sssi_group = sssi2.groupby('Age')
sssi_group = sssi_group.mean()
my_colors = 'rgbkymc'
my_plot = sssi_group.plot(kind='bar')
```

* question - grouping ages?

```python
sssi2 = df[['SSSI_avg','Gender']]
sssi_group = sssi2.groupby('Gender')
sssi_group = sssi_group.mean()
my_colors = 'rgbkymc'
my_plot = sssi_group.plot(kind='bar')
```

* Question - A dataset with categorical variables, like gender or degree, written out?


```python
sssi2 = df[['SSSI_avg','Degree']]
sssi_group = sssi2.groupby('Degree')
sssi_group = sssi_group.mean()
my_colors = 'rgbkymc'
my_plot = sssi_group.plot(kind='bar')
```

## Pie charts for Social Connection Among Asians

```python
slasia1 = df['SLASIA1'].value_counts()
pie1 = pd.DataFrame({'type': slasia1}, 
                     index = [1,2,3,4,5]
                   )
pie1.plot.pie(y='type', figsize=(20,10))
```

## Bar charts for Social Support, with grouping

```python
#take Social Support as the example for bar chart 
support = df[['SocSup1','SocSup2','SocSup3','SocSup4','SocSup5','SocSup6','SocSup7','SocSup8','SocSup9']]
support2 = support.mean()
my_colors = 'rgbkymc'
my_plot = support2.plot(kind='bar')
```

```python
#take Social Support by degree as a bar chart example 
support = df[['SocSup1','SocSup2','SocSup3','SocSup4','SocSup5','SocSup6','SocSup7','SocSup8','SocSup9','Degree']]
support2 = support.groupby(['Degree']).mean()
my_plot = support2.plot(kind='bar',stacked=True,title="Social Support by Degree")
```
