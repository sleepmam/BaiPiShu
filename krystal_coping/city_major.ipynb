```python
import pandas as pd
import numpy as np
```

```python
#import dataset
df = pd.read_csv("location.csv")
df
```

```python
#separate genre types to multiple columns
df.columns
df[['province_before','city_before']] = df.loc_before.str.split("-",expand=True,)
df
```

```python
df_loc = df.set_index('id')
df_loc_expand = df_loc['loc_current'].str.split(expand=True)
df_joined = df_loc.join(df_loc_expand)
df_joined
```

```python
df_clean = df_joined.drop(columns=[2,3,4,5,6,7,8])
df_clean
```

```python
df_clean = df_clean.rename(columns={0: "country_cur", 1: "city_cur"})
df_clean
```

```python

df_major_expand = df_clean['major'].str.split(expand=True)
df_clean2 = df_clean.join(df_major_expand)
df_clean2
```

```python
df_clean2 = df_clean2.drop(columns=[0,2,3,4,5,6,7,8,"major"])
df_clean2
```

```python
df_final = df_clean2.rename(columns={1: "major"})
df_final
```

```python
df_final.to_excel('location_major.xlsx')
```
